## Copy vs View
* Copy is a new array, and the View is just a view of the original array.

```python
arr = np.array([1, 2, 3, 4, 5])
x = arr.copy()
arr[0] = 42

print(arr)
print(x)
```

```python
arr = np.array([1, 2, 3, 4, 5])
x = arr.view()
arr[0] = 42

print(arr)
print(x)
```

# Mesh grid

```python
a = np.linspace(0, 5, 5, endpoint=False)
b = np.linspace(0, 3, 3,  endpoint=False)

print(a)
print(b)
```

```python
xa, xb = np.meshgrid(a, b)
print(xa)
print(xb)
```

```python
import matplotlib.pyplot as plt
plt.scatter(xa,xb, marker = 'o')
plt.show()
```
